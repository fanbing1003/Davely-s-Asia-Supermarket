```python
duplicate = collections.Counter(data["Barcode"])
```

```python
duplicate_barcode = []
for i in duplicate.keys():
  if duplicate[i] > 1:
    duplicate_barcode.append(i)
```

```python
len(duplicate_barcode)
```

```python
duplicate_barcode_item = pd.DataFrame(columns = data.columns)
```

```python
for i in range(len(data)):
  if data.iloc[i,]['Barcode'] in duplicate_barcode:
    duplicate_barcode_item.loc[len(duplicate_barcode_item)] = list(data.iloc[i])
```

```python
len(duplicate_barcode_item)
```
