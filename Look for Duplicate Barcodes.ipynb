```python
import pandas as pd
```

```python
data = pd.read_csv('Data.csv')
```

```python
data.info()
```

```python
char = data[data['Status'] == 'ACTIVE'][['ProductCode', 'Name', 'Barcode']]
```

```python
char = char.reset_index()
char.pop('index')
char
```

```python
len(char['Barcode'].unique())
```

```python
Duplicate_code_list = []
for i in char['Barcode'].unique():
    if len(char[char['Barcode'] == i]) > 1:
        Duplicate_code_list.append(i)
```

```python
len(Duplicate_code_list)
```

### Create another Dataframe that contain all the item which barcode is duplicate

```python
Duplicate = pd.DataFrame(columns = data.columns)
for i in range(len(data)):
    if data['Barcode'][i] in Duplicate_code_list:
        Duplicate.loc[len(Duplicate)] = list(data.loc[i])
```

```python
Duplicate
```

```python
Duplicate['Barcode'].value_counts()
```

```python
Duplicate.to_csv('Duplicate_Barcode_items.csv', index = False)
```
