```python
import pandas as pd
import re
```

```python
data = pd.read_csv('ToScott_190822.csv')
```

```python
len(data['Name'])
```

```python
weight = {}
keys = ['G', 'KG', 'g', 'kg', 'Kg', 'kG']
for i in data['Name']:
  skip = False
  if i in weight:
    skip = True
  if skip == False:
    for num in range(len(i)):
      try:
        test = int(i[num : num+4])#四位數
        for key in keys:
          w = str(test) + key
          if w in i:
            if i not in weight:
              weight[i] = w
              skip = True
            else:
              continue
            break
          else:
            continue
      except:
        continue  
  if skip == False:
    for num in range(len(i)):
      try:
        test = int(i[num : num+3])#三位數
        w = ''
        for key in keys:
          w = str(test) + key
          if w in i:
            if i not in weight:
              weight[i] = w
              skip = True
            else:
              continue
            break
          else:
            continue
      except:
        continue
  if skip == False:
    for num in range(len(i)):
      try:
        test = int(i[num : num+2])#兩位數
        for key in keys:
          w = str(test) + key
          if w in i:
            if i not in weight:
              weight[i] = w
              skip = True
            else:
              continue
            break
      except:
        continue

  if skip == False:
    for num in range(len(i)):
      try:
        test = int(i[num : num+1])#一位數
        for key in keys:
          w = str(test) + key
          if w in i:
            if i not in weight:
              weight[i] = w
              skip = True
            else:
              continue
            break
          else:
            continue

      except:
        continue
  if skip == False:
    weight[i] = 'Missing'
```
