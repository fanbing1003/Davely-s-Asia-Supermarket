## Change the status of the items that are not in the sales history into Deprecate 

```python
Deprecation = pd.DataFrame(columns = data.columns)
```

```python
Deprecation
```

```python
Reserve = pd.DataFrame(columns = data.columns)
```

```python
Reserve
```

```python
Sales_SKU = list(sales_history['SKU'].unique())
for i in range(len(data)):
    if data['ProductCode'][i] not in Sales_SKU:
        Deprecation.loc[len(Deprecation)] = list(data.loc[i])
    else:
        Reserve.loc[len(Reserve)] = list(data.loc[i])
```

## Create a file with active item duplicate barcode

```python
Active = pd.DataFrame(columns = data.columns)
duplicate_barcode = list(Reserve['Barcode'].unique())
for i in duplicate_barcode:
    if len(Reserve[Reserve['Barcode'] == i]) > 1:
        Active = Active.append(Reserve[Reserve['Barcode'] == i])
```

```python
Active = Active.reset_index()
```
